```python
lst = list(df['final_body'])
ids = list(df['Message-ID'])
```

```python
print(len(ids))
print(len(lst))
```

```python
chunk_size = 5
for i in range(0, len(lst), chunk_size):
    chunk_documents = lst[i:i+chunk_size]
    chunk_ids = ids[i:i+chunk_size]
    print(chunk_documents)
    collection.add(
        documents=chunk_documents,
        ids=chunk_ids
    )
```

```python
import numpy as np
model = SentenceTransformer('Alibaba-NLP/gte-large-en-v1.5', trust_remote_code=True).cuda()
# Define chunk size
chunk_size = 5
embeddings = []
# Process the data in chunks
for i in range(0, len(lst), chunk_size):
    chunk_documents = lst[i:i + chunk_size]
    embedding = model.encode(chunk_documents)

    # Append the embeddings to the list
    embeddings.append(embedding)

# Convert list of embeddings to a single NumPy array
embeddings = np.vstack(embeddings)
```
